```python
Dialogos = pd.read_csv('csv/dialogos_completos.csv')
Dialogos
```

## Harry Potter

Vamos a filtrar por el personaje más popular

```python
dialogos_harry_potter = Dialogos[Dialogos['character'] == 'Harry Potter']
dialogos_harry_potter
```

Sacamos mapa de palabras

```python
# filtramos las palabras comunes en inglés
stop_words = set(stopwords.words("english"))

# concatenamos y filtramos las palabras
words = " ".join(dialogos_harry_potter["dialog"])
filtered_words = [word for word in words.split() if word.lower() not in stop_words]

# covertimos las palabras filtradas en un solo texto
text = " ".join(filtered_words)

# creamos el objeto Word Cloud
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)

plt.figure(figsize=(16, 8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
