```python
# vamos a calcularlo sobre 1000 personas:
resultados_6 = {
    "Slytherin_7": 34.54,
    "Ravenclaw_7": 27.71,
    "Griffindor_7": 29.96,
    "Hufflepuff_7": 12.79
}

personas = 1000

resultados_1000_personas_7 = {}

for casa, porcentaje in resultados_6.items():
    resultado_1000 = (porcentaje / 100) * personas
    resultados_1000_personas_7[casa] = resultado_1000

for casa, count in resultados_1000_personas_7.items():
    print(f'{casa}: {int(count)} personas')
```

```python
# hago un superdiccionario con las listas de las casas

data = {
    "Gryffindor": [29, 954, 299, 250],
    "Slytherin": [27, 468, 345, 293],
    "Hufflepuff": [12, 198, 127, 160],
    "Ravenclaw": [10, 288, 277, 272]
}
```

```python
df = pd.DataFrame(data) # lo paso a dataframe
df
```

```python
sum_df = pd.DataFrame(df.sum(), columns=["Suma Total"]) # ok, pero no me convence
sum_df
```

```python
transposed_df = df.T # así vamos por mejor camino
transposed_df
```

```python
transposed_df["Total"] = transposed_df.iloc[:, 1:].sum(axis=1)
transposed_df
```

```python
transposed_df = transposed_df.sort_values(by="Total", ascending=False) # ordeno por total de mayor a menor
transposed_df
```

```python
transposed_df["Orden"] = list(range(1, len(transposed_df) + 1)) # añado orden
transposed_df
```

```python
contador_column = transposed_df.pop("Orden")
transposed_df.insert(0, "Orden", contador_column) # lo ponemos en primer lugar y listo!!!!
transposed_df
```

```python
transposed_df.to_csv('csv/casas.csv', index=False) # paso a csv
```
