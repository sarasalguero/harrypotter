```python
import numpy as np
import pandas as pd
```

```python
df_1 = pd.read_csv('data/Spells.csv')

df_2 = pd.read_csv('data/Harry_Potter_Spells.csv', sep = ";")
```

```python
df_1.head()
```

```python
df_2.head() # tengo que cambiar el nombre de la columna de los echizos
```

```python
df_2 = df_2.rename(columns={'Name of the spell': 'Incantation'}) # cambio el nombre
df_2.head()
```

```python
# filtro para quedarme solo con las que me interesan
columnas_deseadas = ['Incantation', 'Complexity (from 1 to 7 year)']
df_2_filtrado = df_2[columnas_deseadas]
df_2_filtrado
```

```python
merged_df = pd.merge(df_1, df_2_filtrado, on='Incantation', how='left') # uno ambos dataframes
merged_df
```

```python
# cambio los NaN en las luces por un "No color" porque no es que falte el dato, es que esos hechizos no se iluminan.
merged_df['Light'] = merged_df['Light'].fillna('No color')
merged_df
```

```python
merged_df.info() # todo parece estar ok, solo hay NaN en complexity pero era de esperar
```

```python
# Saco el csv a una carpeta nueva donde guardaré todos los csv nuevos que haga
merged_df.to_csv('csv/hechizos.csv', index=False)
```
